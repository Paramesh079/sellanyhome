```python
# --- Column Classification ---
target_col = 'price(EAD)'

date_cols = ['date']

num_cols = ['built_up(sqft)', 'plot(sqft)']

cat_cols = ['main_location', 'sub_location', 'building_project',
            'is_off_plan?', 'is_Vacant_at_time_of_sale?', 'type', 'beds']

cat_cols = [c for c in cat_cols if c in df.columns]
num_cols = [c for c in num_cols if c in df.columns]

print(f'🎯 Target:      {target_col}')
print(f'📅 Date cols:    {date_cols}')
print(f'🔢 Num cols:     {num_cols}')
print(f'🏷️  Cat cols:     {cat_cols}')
```

```python
# Remove duplicates based on all columns
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)

print(f'Before: {before:,} rows')
print(f'Removed: {before - after:,} duplicates')
print(f'After:  {after:,} rows')
```

```python
# --- Convert columns to proper data types ---

# Target → float
df[target_col] = df[target_col].astype(str).str.replace(',', '').str.strip()
df[target_col] = pd.to_numeric(df[target_col], errors='coerce')

# Numeric cols → float (handle '-' and commas)
for col in num_cols:
    df[col] = df[col].astype(str).str.replace(',', '').str.strip()
    df[col] = df[col].replace(['-', '', 'nan', 'None'], np.nan)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Date cols → datetime
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed', dayfirst=True, errors='coerce')

# Categorical cols → category
for col in cat_cols:
    df[col] = df[col].astype('category')

print('✅ Data types converted!\n')
print(df.dtypes)
```

```python
# --- Rows with any null values ---
null_rows = df[df.isnull().any(axis=1)]
print(f'Rows with at least one null: {len(null_rows):,} / {len(df):,} ({len(null_rows)/len(df)*100:.2f}%)\n')

# Show which columns have nulls in these rows
print('Null count per column (in null rows only):')
print(null_rows.isnull().sum().sort_values(ascending=False))

```

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 1. Create Dropdowns
w_main = widgets.Dropdown(description='Location:')
w_sub = widgets.Dropdown(description='Sub-Loc:')
w_build = widgets.Dropdown(description='Building:')

out = widgets.Output()

def update_sub(*args):
    """Update sub-location dropdown based on chosen main location"""
    if w_main.value:
        subs = df[df['main_location'] == w_main.value]['sub_location'].dropna().unique()
        w_sub.options = sorted(subs)
        if len(w_sub.options) > 0:
            w_sub.value = w_sub.options[0]

def update_build(*args):
    """Update building dropdown based on chosen sub-location"""
    if w_main.value and w_sub.value:
        builds = df[(df['main_location'] == w_main.value) & 
                    (df['sub_location'] == w_sub.value)]['building_project'].dropna().unique()
        w_build.options = sorted(builds)
        if len(w_build.options) > 0:
            w_build.value = w_build.options[0]

def update_plot(*args):
    with out:
        out.clear_output(wait=True)
        
        # Filter the dataset
        subset = df[(df['main_location'] == w_main.value) & 
                    (df['sub_location'] == w_sub.value) & 
                    (df['building_project'] == w_build.value)].copy()
        
        if subset.empty:
            print("⚠️ No data available for this selection.")
            return

        # Handle formatting: remove commas and convert to numeric
        subset['price(EAD)'] = subset['price(EAD)'].astype(str).str.replace(',', '').str.strip()
        subset['price(EAD)'] = pd.to_numeric(subset['price(EAD)'], errors='coerce')
        
        subset['built_up(sqft)'] = subset['built_up(sqft)'].astype(str).str.replace(',', '').str.strip()
        subset['built_up(sqft)'] = pd.to_numeric(subset['built_up(sqft)'], errors='coerce')
        
        # If plot area exists, clean it too
        if 'plot(sqft)' in subset.columns:
            subset['plot(sqft)'] = subset['plot(sqft)'].astype(str).str.replace(',', '').str.strip()
            subset['plot(sqft)'] = pd.to_numeric(subset['plot(sqft)'], errors='coerce')
        
        subset['date'] = pd.to_datetime(subset['date'], format='mixed', dayfirst=True, errors='coerce')
        
        # Drop rows missing crucial time/price data & sort by date
        subset = subset.dropna(subset=['price(EAD)', 'date'])
        subset = subset.sort_values(by='date')
        
        if subset.empty:
            print("⚠️ No valid price/date data available for this selection.")
            return
            
        # --- Analytics Calculations ---
        count = len(subset)
        max_val = subset['price(EAD)'].max()
        min_val = subset['price(EAD)'].min()
        off_plan_count = subset['is_off_plan?'].astype(str).str.lower().isin(['true', '1', 'yes']).sum()
        
        # Print Analytics Summary
        print("📊 Analytics Summary")
        print("-" * 45)
        print(f"Total Transactions  : {count:,}")
        print(f"Off-Plan Sales      : {off_plan_count:,} ({ (off_plan_count/count)*100:.1f}%)")
        print(f"Max Price           : {max_val:,.0f} AED")
        print(f"Min Price           : {min_val:,.0f} AED")
        print("-" * 45)
        
        # ==========================================
        # Plot : Price Pattern Over Time
        # ==========================================
        plt.figure(figsize=(14, 6))
        
        group_col = 'beds' if 'beds' in subset.columns else 'type'
        categories = subset[group_col].dropna().unique()
        colors = plt.cm.tab10(np.linspace(0, 1, max(1, len(categories))))
        
        for idx, cat in enumerate(categories):
            cat_data = subset[subset[group_col] == cat]
            
            # Draw line
            plt.plot(cat_data['date'], cat_data['price(EAD)'], '-', color=colors[idx], alpha=0.4, linewidth=1.5)
            
            # Draw uniform points
            plt.scatter(cat_data['date'], cat_data['price(EAD)'], 
                        s=60, color=colors[idx], alpha=0.7, edgecolors='black', 
                        label=f"{cat} {group_col.title()}")

        plt.title(f"Price Pattern Over Time - {w_build.value}", fontsize=15, fontweight='bold')
        plt.xlabel("Date", fontsize=12)
        plt.ylabel("Price (AED)", fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.xticks(rotation=45)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Unit Details")
        plt.tight_layout()
        plt.show()

# 3. Attach observers
w_main.observe(update_sub, 'value')
w_sub.observe(update_build, 'value')
w_build.observe(update_plot, 'value')

# 4. Initialize first values
main_locs = sorted(df['main_location'].astype(str).unique())
w_main.options = [loc for loc in main_locs if loc != 'nan']

if len(w_main.options) > 0:
    w_main.value = w_main.options[0]
    update_sub()
    update_build()
    update_plot()

# 5. Display the UI
ui = widgets.VBox([
    widgets.HBox([w_main, w_sub, w_build]), 
    out
])
display(ui)
```

## price per sqft distribution

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# --- Ensure price_per_sqft exists for later calculations ---
if 'price_per_sqft' not in df.columns:
    df['price_per_sqft'] = pd.to_numeric(df['price(EAD)'].astype(str).str.replace(',', '').str.strip(), errors='coerce') / \
                           pd.to_numeric(df['built_up(sqft)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

# 1. Create Dropdowns
w_main = widgets.Dropdown(description='Location:')
w_sub = widgets.Dropdown(description='Sub-Loc:')
w_build = widgets.Dropdown(description='Building:')

out = widgets.Output()

def update_sub(*args):
    """Update sub-location dropdown based on chosen main location"""
    if w_main.value:
        subs = df[df['main_location'] == w_main.value]['sub_location'].dropna().unique()
        w_sub.options = ['All'] + sorted(subs)
        w_sub.value = w_sub.options[0]

def update_build(*args):
    """Update building dropdown based on chosen sub-location"""
    if w_main.value and w_sub.value != 'All':
        builds = df[(df['main_location'] == w_main.value) & 
                    (df['sub_location'] == w_sub.value)]['building_project'].dropna().unique()
        w_build.options = ['All'] + sorted(builds)
        w_build.value = w_build.options[0]
    elif w_main.value and w_sub.value == 'All':
        builds = df[(df['main_location'] == w_main.value)]['building_project'].dropna().unique()
        w_build.options = ['All'] + sorted(builds)
        w_build.value = w_build.options[0]

def plot_distribution(axis, data_series, title, color):
    """Helper function to draw a single distribution plot"""
    plot_data = data_series.dropna()
    if not plot_data.empty:
        sns.histplot(plot_data, kde=True, bins=30, color=color, ax=axis)
        median_val = plot_data.median()
        axis.axvline(median_val, color='red', linestyle='--', label=f'Median: {median_val:,.0f}')
        axis.set_title(title, fontsize=13, fontweight='bold')
        axis.set_xlabel('Price per Sqft (AED)', fontsize=11)
        axis.set_ylabel('Frequency', fontsize=11)
        axis.legend()
    else:
        axis.text(0.5, 0.5, 'No valid data available', 
                 horizontalalignment='center', verticalalignment='center', transform=axis.transAxes,
                 fontsize=12, color='gray')
        axis.axis('off')

def update_plot(*args):
    with out:
        out.clear_output(wait=True)
        
        # Datasets for each level
        df_main = df[df['main_location'] == w_main.value].copy()
        df_sub = pd.DataFrame()
        df_build = pd.DataFrame()
        
        if w_sub.value != 'All':
            df_sub = df_main[df_main['sub_location'] == w_sub.value].copy()
            
        if w_build.value != 'All':
            # Handle if Sub == All vs Sub == Specific
            if df_sub.empty:
                df_build = df_main[df_main['building_project'] == w_build.value].copy()
            else:
                df_build = df_sub[df_sub['building_project'] == w_build.value].copy()
        
        # Layout depends on how many specific filters are applied
        plots_to_draw = 1
        if w_sub.value != 'All': plots_to_draw += 1
        if w_build.value != 'All': plots_to_draw += 1
            
        fig, axes = plt.subplots(plots_to_draw, 1, figsize=(14, 6 * plots_to_draw))
        if plots_to_draw == 1:
            axes = [axes] # Make it iterable
            
        current_ax_idx = 0
        
        # 1. Plot Main Location
        plot_distribution(axes[current_ax_idx], df_main['price_per_sqft'], f'Main Location - {w_main.value}', '#3498db')
        current_ax_idx += 1
        
        # 2. Plot Sub Location (if selected)
        if not df_sub.empty:
            plot_distribution(axes[current_ax_idx], df_sub['price_per_sqft'], f'Sub Location - {w_sub.value}', '#e67e22')
            current_ax_idx += 1
            
        # 3. Plot Building (if selected)
        if not df_build.empty:
            plot_distribution(axes[current_ax_idx], df_build['price_per_sqft'], f'Building - {w_build.value}', '#8e44ad')

        plt.tight_layout(pad=3.0)
        plt.show()

# 3. Attach observers
w_main.observe(update_sub, 'value')
w_sub.observe(update_build, 'value')
w_main.observe(update_plot, 'value')
w_sub.observe(update_plot, 'value')
w_build.observe(update_plot, 'value')

# 4. Initialize first values
main_locs = sorted(df['main_location'].astype(str).unique())
w_main.options = [loc for loc in main_locs if loc != 'nan']

if len(w_main.options) > 0:
    w_main.value = w_main.options[0]
    update_sub()
    update_build()
    update_plot()

# 5. Display the UI
ui = widgets.VBox([
    widgets.HBox([w_main, w_sub, w_build]), 
    out
])
display(ui)
```

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 1. Create Dropdowns
w_main = widgets.Dropdown(description='Location:')
w_sub = widgets.Dropdown(description='Sub-Loc:')
w_build = widgets.Dropdown(description='Building:')

out = widgets.Output()

def update_sub(*args):
    """Update sub-location dropdown based on chosen main location"""
    if w_main.value:
        subs = df[df['main_location'] == w_main.value]['sub_location'].dropna().unique()
        w_sub.options = sorted(subs)
        if len(w_sub.options) > 0:
            w_sub.value = w_sub.options[0]

def update_build(*args):
    """Update building dropdown based on chosen sub-location"""
    if w_main.value and w_sub.value:
        builds = df[(df['main_location'] == w_main.value) & 
                    (df['sub_location'] == w_sub.value)]['building_project'].dropna().unique()
        w_build.options = sorted(builds)
        if len(w_build.options) > 0:
            w_build.value = w_build.options[0]

def update_plot(*args):
    with out:
        out.clear_output(wait=True)
        
        # Filter the dataset
        subset = df[(df['main_location'] == w_main.value) & 
                    (df['sub_location'] == w_sub.value) & 
                    (df['building_project'] == w_build.value)].copy()
        
        if subset.empty:
            print("⚠️ No data available for this selection.")
            return

        # Handle formatting: remove commas and convert to numeric
        subset['price(EAD)'] = subset['price(EAD)'].astype(str).str.replace(',', '').str.strip()
        subset['price(EAD)'] = pd.to_numeric(subset['price(EAD)'], errors='coerce')
        
        subset['built_up(sqft)'] = subset['built_up(sqft)'].astype(str).str.replace(',', '').str.strip()
        subset['built_up(sqft)'] = pd.to_numeric(subset['built_up(sqft)'], errors='coerce')
        
        # Ensure plot(sqft) exists and is numeric
        if 'plot(sqft)' in subset.columns:
            subset['plot(sqft)'] = subset['plot(sqft)'].astype(str).str.replace(',', '').str.strip()
            subset['plot(sqft)'] = pd.to_numeric(subset['plot(sqft)'], errors='coerce')
        else:
            subset['plot(sqft)'] = np.nan
        
        subset['date'] = pd.to_datetime(subset['date'], format='mixed', dayfirst=True, errors='coerce')
        
        # Drop rows missing crucial time/price data & sort by date
        subset = subset.dropna(subset=['price(EAD)', 'date'])
        subset = subset.sort_values(by='date')
        
        if subset.empty:
            print("⚠️ No valid price/date data available for this selection.")
            return
            
        # --- Basic Analytics Calculations ---
        count = len(subset)
        max_val = subset['price(EAD)'].max()
        min_val = subset['price(EAD)'].min()
        off_plan_count = subset['is_off_plan?'].astype(str).str.lower().isin(['true', '1', 'yes']).sum()
        
        # --- Built-Up Price per Sqft Calculations ---
        subset['price_per_sqft_builtup'] = subset['price(EAD)'] / subset['built_up(sqft)']
        valid_bu_ppsf = subset['price_per_sqft_builtup'].dropna()
        bu_median = valid_bu_ppsf.median() if not valid_bu_ppsf.empty else 0
        bu_mean = valid_bu_ppsf.mean() if not valid_bu_ppsf.empty else 0
        
        # --- Plot Area Price per Sqft Calculations ---
        subset['price_per_sqft_plot'] = subset['price(EAD)'] / subset['plot(sqft)']
        valid_plot_ppsf = subset['price_per_sqft_plot'].dropna()
        plot_median = valid_plot_ppsf.median() if not valid_plot_ppsf.empty else 0
        plot_mean = valid_plot_ppsf.mean() if not valid_plot_ppsf.empty else 0

        # === Determine Primary Property Type ===
        # We find the most common property type in the selection
        dominant_type = subset['type'].mode()[0].lower() if not subset['type'].empty else ""
        
        # Determine classification flags based on the type name
        is_villa_type = 'villa' in dominant_type or 'townhouse' in dominant_type
        is_plot_type = 'plot' in dominant_type

        # Print Analytics Summary
        print("📊 Analytics Summary for:", w_build.value)
        print("-" * 65)
        print(f"Total Transactions  : {count:,}")
        print(f"Off-Plan Sales      : {off_plan_count:,} ({ (off_plan_count/count)*100:.1f}%)")
        print(f"Max Price           : {max_val:,.0f} AED")
        print(f"Min Price           : {min_val:,.0f} AED")
        print("-" * 65)
        
        if is_villa_type and not valid_bu_ppsf.empty and not valid_plot_ppsf.empty:
            # IT'S A VILLA/TOWNHOUSE (Display both)
            print(f"Calculated for Type: {dominant_type.title()}")
            print(f"\n{'METRIC':<25} | {'BUILT-UP AREA':<18} | {'PLOT AREA':<18}")
            print("-" * 65)
            print(f"{'Median Price/Sqft':<25} | AED {bu_median:<14,.0f} | AED {plot_median:<14,.0f}")
            print(f"{'Average Price/Sqft':<25} | AED {bu_mean:<14,.0f} | AED {plot_mean:<14,.0f}")
            
        elif is_plot_type and not valid_plot_ppsf.empty:
            # IT'S A PLOT OF LAND (Only plot space)
            print(f"Calculated for Type: {dominant_type.title()} (Based on Plot Area)")
            print(f"Median Price/Sqft   : AED {plot_median:,.0f}")
            print(f"Average Price/Sqft  : AED {plot_mean:,.0f}")
            
        elif not valid_bu_ppsf.empty:
            # DEFAULT/APARTMENT/HOTEL/PENTHOUSE (Only built-up space)
            print(f"Calculated for Type: {dominant_type.title()} (Based on Built-Up Area)")
            print(f"Median Price/Sqft   : AED {bu_median:,.0f}")
            print(f"Average Price/Sqft  : AED {bu_mean:,.0f}")
            
        else:
            print(f"Primary Type: {dominant_type.title()}")
            print("⚠️ Insufficient area data to compute Price/Sqft metrics.")
            
        print("-" * 65)
        
        # ==========================================
        # Plot 1 : Price Pattern Over Time
        # ==========================================
        plt.figure(figsize=(14, 6))
        
        group_col = 'beds' if 'beds' in subset.columns else 'type'
        categories = subset[group_col].dropna().unique()
        colors = plt.cm.tab10(np.linspace(0, 1, max(1, len(categories))))
        
        for idx, cat in enumerate(categories):
            cat_data = subset[subset[group_col] == cat]
            
            # Draw line
            plt.plot(cat_data['date'], cat_data['price(EAD)'], '-', color=colors[idx], alpha=0.4, linewidth=1.5)
            
            # Draw uniform points
            plt.scatter(cat_data['date'], cat_data['price(EAD)'], 
                        s=60, color=colors[idx], alpha=0.7, edgecolors='black', 
                        label=f"{cat} {group_col.title()}")

        plt.title(f"Price Pattern Over Time - {w_build.value}", fontsize=15, fontweight='bold')
        plt.xlabel("Date", fontsize=12)
        plt.ylabel("Price (AED)", fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.xticks(rotation=45)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Unit Details")
        plt.tight_layout()
        plt.show()

# 3. Attach observers
w_main.observe(update_sub, 'value')
w_sub.observe(update_build, 'value')
w_build.observe(update_plot, 'value')

# 4. Initialize first values
main_locs = sorted(df['main_location'].astype(str).unique())
w_main.options = [loc for loc in main_locs if loc != 'nan']

if len(w_main.options) > 0:
    w_main.value = w_main.options[0]
    update_sub()
    update_build()
    update_plot()

# 5. Display the UI
ui = widgets.VBox([
    widgets.HBox([w_main, w_sub, w_build]), 
    out
])
display(ui)
```
